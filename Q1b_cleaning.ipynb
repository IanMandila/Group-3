```python
import pandas as pd
import numpy as np
```

```python
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'race', 'sex',
    'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income'
]
```

```python
df = pd.read_csv(
    'data/adult.data',
    names=columns,
    sep=',',
    skipinitialspace=True,
    na_values='?'
)

df.head()
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip().str.lower()
```

```python
df['workclass'] = df['workclass'].fillna('unknown')
df['occupation'] = df['occupation'].fillna('unknown')
df['native_country'] = df['native_country'].fillna('unknown')
```

```python
df = df.drop(columns=['education'])
```

```python
# 1. Encode the target first
df['income'] = df['income'].map({'<=50k': 0, '>50k': 1})
```

```python
# 2. Then one-hot encode only the categorical features
df_encoded = pd.get_dummies(df.drop(columns=['income']), drop_first=True)
```

```python
# 3. Add the target back
df_encoded['income'] = df['income']
```

```python
df_encoded.to_csv('adult_cleaned.csv', index=False)
```
